# Feature Engineering

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)
```

```python
# since order date and ship date were in categorical form instead of dt format, using above code it gets converted to dt format.
```

```python
df.info()
```

```python
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
# dt.year extract the year from date.
# dt.month extracts the month from date.
# shipping date is the newly created column in which dt.days extracts the total no. of days required for shipping.
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
